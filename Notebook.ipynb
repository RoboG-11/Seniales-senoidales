# Explicación de las propiedades de las señales senoidales

### Introducción

Las señales senoidales o señales sinusoidales son patrones de ondas periodicas que matemáticamente pueden ser descritas mediante las funciones del seno y coseno. Estas describen acertadamente eventos naturales y señales variables en el tiempo.  
Este tipo de señales están compuestas por varias partes, las cuales se explicarán más adelante.

Se puede representar como:  
$y(t)=Asin(2πft+φ)$

Donde:
- $A$. Amplitud: indica el valor máximo a la que la senoide va a llegar. 
- $f$. Frecuencia: representa el ángulo girado por unidad de tiempo.
- $t$. Periodo: es la cantidad de tiempo en el que se acompleta un ciclo. *Esta es la variable independiente*.
- $φ$. Fase: es el desplazamiento a la izquiero o derecha.


[fuente](https://www.lifeder.com/onda-senoidal/)

```python
import matplotlib.pyplot as plt
import numpy as np

#* Variables que determinan los valores de la señal *#
amplitud = 1
frecuencia = 1
periodo = 1 / frecuencia
fase = 1

#* Frecuencia, periodo de muestreo y ciclos *#
frecuencia_Muestreo = 30 * frecuencia
periodo_Muestreo = 1 / frecuencia_Muestreo
ciclos = 5

t = np.arange( 0, ciclos * periodo, periodo_Muestreo ) # Variable independiente
x = amplitud * np.sin( 2 * np.pi * frecuencia * t + fase ) # Caulculo de la señal

#* Graficación de la señal *#
plt.axhline(y = 0, color = 'k')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.plot(t,x)
plt.grid('on')
```

# AMPLITUD

La amplitud de una señal senoidal es la distancia que hay desde la línea central hasta la parte superior de una cresta o al fondo de un canal.  
En el siguiente fragmento de código puede cambiar el valor de la variable *amplitud*, esto para que se pueda visualizar con mayor claridad este concepto.


[fuente](https://support.apple.com/es-mx/guide/mainstage/lgsife418f71/mac#:~:text=Amplitud%3A%20la%20amplitud%20de%20una,l%C3%ADnea%20horizontal%20a%200%20dB)

```python
#* Variable a modificar *#
amplitud = 4

t = np.arange( 0, ciclos * periodo, periodo_Muestreo ) # Variable independiente**
x = amplitud * np.sin( 2 * np.pi * frecuencia * t + fase ) # Caulculo de la señal

#* Graficación de la señal *#
plt.axhline(y = 0, color = 'k')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.plot(t,x)
plt.grid('on')
```

# Señal compuesta
Una señal compuesta está compuesta por varias señales senoidales de diferentes aplitutes, frecuencias y fases. Pueden ser periódica o no periódica. Una señal compuesta periódica puede descomponerse en una serie de ondas senoidales simples con frecuencias discretas, frecuencias que tienen valores enteros. Mientras que una señal compuesta no periódica puede descomponerse en una combinación de un número infinito de ondas sinusoidales simples con frecuencias continuas, frecuencias que tienen valores reales.  
Un ejemplo de este tipo de señales es una señal de video, que tiene dos señales de audio y una señal de imagen.
