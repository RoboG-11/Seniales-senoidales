# Explicación de las propiedades de las señales senoidales

### Introducción

Las señales senoidales o señales sinusoidales son patrones de ondas periodicas que matemáticamente pueden ser descritas mediante las funciones del seno y coseno. Estas describen acertadamente eventos naturales y señales variables en el tiempo.  
Este tipo de señales están compuestas por varias partes, las cuales se explicarán más adelante.

Las señales senoidales se puede representar de la siguiente manera:  
$y(t)=Asin(2πft+φ)$

Donde:
- $A$. Amplitud: indica el valor máximo a la que la senoide va a llegar. 
- $f$. Frecuencia: representa el ángulo girado por unidad de tiempo.
- $t$. Tiempo: es la cantidad de tiempo en el que se acompleta un ciclo. *Esta es la variable independiente*.
- $φ$. Fase: es el desplazamiento a la izquiero o derecha.

```python
#* Bibliotecas utilizadad *#
import matplotlib.pyplot as plt
import numpy as np

#* Variables que determinan los valores de la señal *#
amplitud = 1
frecuencia = 1
periodo = 1 / frecuencia
fase = 1

#* Frecuencia, periodo de muestreo y ciclos *#
frecuencia_Muestreo = 30 * frecuencia
periodo_Muestreo = 1 / frecuencia_Muestreo
ciclos = 5

t = np.arange( 0, ciclos * periodo, periodo_Muestreo ) # Variable independiente
x = amplitud * np.sin( 2 * np.pi * frecuencia * t + fase ) # Caulculo de la señal

#* Graficación de la señal *#
plt.axhline(y = 0, color = 'k')
plt.title('Señal senoidal')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.grid('on')
plt.plot(t,x)
```

# Explicación de una señal periódica

Las señales periódicas se caracterizan porque se les puede encontar un patrón que se repite constantemente cada determinado tiempo. El **periodo** es el tiempo que tarda una función en completar un ciclo y este se expresa en segundos. 

Para comprender porque una señal senoidal se considera periódica podríamos imaginar un plano y una circunferencia con su centro en el origen y radio de 1 unidad. Ahora supongamos que tenemos un punto móvil que se desplaza por el perimetro de la cirunferencia, y además, se traza un segmento desde el origen a este punto, de esta forma se genera un ángulo entre el radio y este segmento. Si calculamos el seno del ángulo mientras que desplazamos el punto móvil nos daremos cuenta que el resultado del seno se repetirá cada que se recorra el circulo completo.

Para poder visualizar el periodo en la gráfica, si el parametro **_"period"_** de la señal senoidal se modifica a 1, la forma de la señal se repetirá cada segundo, si se modifica a 2 se repetirá cada 2 segundos. Inténtalo.

Fuente: [unet.edu.ve](http://www.unet.edu.ve/aula10c/Asenales/Unid01/seg01.htm)

<center><img src=https://www.universoformulas.com/imagenes/matematicas/trigonometria/grafica-seno.jpg></center>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)


period=6.
x_max = 2*pi
x_min = 0
step = 0.01

# Configuración del gráfico visual
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [sine_wave(1,1/period, 0, time) for time in x]
plt.title('Señal senoidal con periodo modificable')
plt.plot(x,y)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.grid('on')
a =[0,period]
b =[0,0]
plt.plot(a, b, linestyle="-", marker="o")
plt.text(period-1, .025, 'Periodo', color='orange')
plt.show()
```

# Frecuencia 

Esta idea coincide con el concepto de periodo de una función periódica, como vimos, la función $f(x)= sen x$ es periódica de periodo $2π$ y si consideras $x0 ∈ R$ y el punto $(x0, senx0)$ en su gráfica, como una partícula moviéndose a lo largo de la curva. 
Ahora bien, se sabe  que la partícula, en su movimiento a lo largo de un periodo, pasa por todos los posibles valores de la función, es decir completa un ciclo. Al número de ciclos completados en el intervalo $[0,2π]$ se le llama velocidad angular, y se denota por la letra griega $ω$. O sea que la velocidad angular es el número de veces que se repite un patrón completo de una curva senoidal en el intervalo $[0,2π]$, en la expresión general de la familia de funciones senoidales, corresponde al parámetro b. También se le conoce como pulsación (o simplemente pulso). 
De acuerdo con la definición de periodo, se tiene que $T= 2π/ω$
Definimos la frecuencia de oscilación, $fr$, de una función senoidal como el recíproco del periodo, es decir $fr = ω/2π = 1/T$
Sucede entonces que si la frecuencia es mayor, el periodo disminuye y viceversa.Su unidad es el hertzio, de símbolo Hz.

Para visualizar el concepto de frecuencia se realizo la grafica de dos señales senoidales. La primera señal esta definida por una frecuencia de valor 4, esta dada por la señal de color azul. La segunda grafica esta definida por una frecuencia de 10, esta dada por la señal de color naranja.Lo que se puede apreciar es que la segunda señal realiza un mayor numero de ciclos completados en un 1s o en el intervalo [0,2π]. 
Se puede modificar  el valor de las frecuencias para poder apreciar su comportamiento, el valor de la frecuencia de la primera señal lo tiene la variable **_"frecuencia1"_** y la de la segunda señal esta dada por la variable frecuencia2, estos valores se pueden modificar con cualquier valor, siempre y cuando la **_"frecuencia2"_** sea mayor a frecuencia1

Fuente: [uapas1.bunam.unam.mx](http://uapas1.bunam.unam.mx/matematicas/funcion_senoidal_b/)

```python
import numpy as np
import matplotlib.pyplot as plt
print("Frecuencia")
#señal1 
frecuencia1 = 4
Periodo1= 1/frecuencia1
A1= 2
theta1 = 0
#señal2
frecuencia2 = 10 #Tiene que ser mayor que la frecuencia1
Periodo2= 1/frecuencia2
A2= 2
theta2 = 0

Fs = 30*frecuencia2
Ts = 1/Fs
Nc= 4
plt.title("Señal senoidal con frecuencia modificable")
t = np.arange(0, Nc*Periodo1, Ts)
x = A1*np.sin(2*np.pi*frecuencia1*t+theta1)
x2 = A2*np.sin(2*np.pi*frecuencia2*t+theta2)
plt.plot(t,x,t,x2)
plt.grid("on")
```

# Explicación del concepto de fase

Para que 2 ondas senoidales se encuentren en fase, deben coincidir las crestas y valles, por lo que la fase hace referencia a la diferencia de tiempo entre dos ondas senoidales, cabe mencionar que no se mide en términos de tiempo sino que lo hace en terminos de ángulo ya sea en grados o radianes.

Además, el **desplazamiento de fase**, hace referencia a la diferencia de fase entre 2 ondas. Por ejemplo, un desplazamiento de 180 grados es un desplazamiento de 1/2 del periodo de la onda. El desplazamiento puede ser positivo o negativo, lo que quiere decir que una onda puede atrasada o avanzada a otra. 

Para poder visualizar el desfase de 2 señales senoidales en la gráfica, si el parametro **_"phase"_** de la señal senoidal se modifica a pi/4 la onda tiene un desplazamiento de 1/8 del periodo, si se modifica el valor a pi tiene un desplazamiento de 1/2. Inténtalo.

Fuente: [lucianonelli.files.wordpress.com](https://lucianonelli.files.wordpress.com/2013/06/las-fases-de-los-sinusoides.pdf)

![hola](https://raw.githubusercontent.com/georgeif2212/Seniales-senoidales/main/IMG/radianes.png)

```python
def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

def sine_wave_phase(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

phase = pi
x_max = 2*pi
x_min = 0
step = 0.01

# Señal 1
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [sine_wave(1,1/2, 0, time) for time in x]
plt.plot(x,y)
# Señal 2 desfasada
a = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
b = [sine_wave_phase(1,1/2, phase, time) for time in x]
plt.plot(a,b)
# Configuración visual
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.grid('on')
plt.show()
```

# Longitud de Onda 

La longitud de onda , distancia entre los puntos correspondientes de dos consecutivos “Puntos correspondientes” se refiere a dos puntos o partículas en la misma fase, es decir, puntos que han completado fracciones idénticas de su movimiento periódico . Por lo general, en ondas transversales (ondas con puntos que oscilan en ángulo recto con respecto a la dirección de su avance), la longitud de onda se mide de cresta a cresta o de valle a valle; en ondas longitudinales (ondas cuyos puntos vibran en la misma dirección que su avance), se mide de compresión en compresión o de rarefacción en rarefacción. La longitud de onda generalmente se denota con la letra griega lambda $(λ)$; es igual a la velocidad $( v ) $ de un tren de ondas en un medio dividida por su frecuencia $( f ): λ = v / f$ .

# Señal compuesta

# Explicación de la importancia de las señales senoidales y las señales compuestas

La modulación requiere de dos señales diferentes, una señal portadora y una moduladora. La primera se trata de una onda de alta frecuencia a la cual se le puede modificar su amplitud, frecuencia o fase, mientras que la segunda hace referencia a la información o datos que queremos transmitir. 
El *indice de modulación* o también llamada *profundidad de modulación* hace referencia a una medida de la variación de amplitud de una onda portadora.
- Modulación AM: Menor igual a 1
- Sobremodulación: se produce cuando la profundidad de modulación es mayor a 1, da lugar a bandas laterales adicionales que causan interferencias llamadas "salpicaduras" al espectro cercano

Estas señales son importantes porque son altamente utilizadas en la vida diaria con múltiples usos en los medios de comunicación, por ejemplo la amplitud modulada se ocupa para poder escuchar la radio, mientras que las señales compuestas se utilizan para mostrar videos y combinar señales de audio y video.

En el siguiente celda de código, si modificas la variable **_Fp_** cambiarás la frecuencia de la señal portadora, **_Fm_** modifica la frecuencia de la señal moduladora, **_Im_** modifica el índice de modulación y **_Amp_** la amplitud de la señal portadora. Intentálo.

Fuente: [eduvirtual.cuc.edu.co](https://eduvirtual.cuc.edu.co/moodle/pluginfile.php/246068/mod_resource/content/1/Practica%20-%20Modulacion%20en%20Amplitud.pdf)

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

plt.close('all')

# m = Valor pico max -Valor picomin /Valor max + Valor min 

#Eje del tiempo (x)
Fs = 2000;
n = np.arange(0,10,1/Fs)

# Señal portadora de alta frecuencia
Fp = 1000 # Frecuencia 
Amp = 1 # Amplitud
c = Amp*np.cos(2*pi*Fp*n) 

# Señal moduladora 
Fm = 1/2 # Frecuencia
Im = 0.9 # Profundidad de modulación AM 
m = Im*np.sin(2*pi*Fm*n)

#Señal modulada por amplitud
s= c * (1 + m/Amp)

plt.plot(n,s); 
plt.title('Modulación de señal')
plt.xlabel('Segundos'); 
plt.ylabel('Amplitud');
plt.grid(True)
```

# Ancho de Banda 

El ancho de banda de una señal se define como la diferencia (usualmente expresada en Hz) entre las frecuencias mínima, fmin, y máxima, fmax, nominales. El problema surge al definir dichas frecuencias, dando lugar a diferentes definiciones para este concepto. Las más utilizadas son las siguientes: 

1.-El ancho de banda absoluto, fmax-fmin, donde el espectro de la señal es nulo fuera del intervalo delimitado por ambas frecuencias. 
2.-El ancho de banda de 3dB o ancho de banda de potencia mitad, fmax-fmin, donde estas frecuencias se fijan de forma que las componentes que quedan fuera del intervalo estén 3dB por debajo de la componente espectral de magnitud mayor.           
3.-El ancho de banda de potencia, fmax-fmin, donde estas frecuencias definen la banda en la que reside el 99% de al potencia total de la señal.                                                                                                 
4.-El ancho de banda nulo a nulo o ancho de banda de cruce cero, fmax-fmin, donde estas frecuencias definen el primer nulo del espectro a cada lado de la componente espectral de mayor magnitud.

# Transformada de Fourier 
Una de las suposiciones basicas al aplicar el metodos de fourier es que la señal oscilatoria se puede descomponer en un montón de señales sinusoidales. La idea  Invirtiendo esa lógica, cualquier señal periódica puede reconstruirse a partir de una serie de sinusoides. Por lo tanto, cualquier señal periódica puede representarse de manera equivalente mediante una serie sinusoidal.básica es que cualquier señal puede representarse como una suma ponderada de ondas seno y coseno de diferentes frecuencias.
La técnica para determinar la representación en serie sinusoidal de una señal periódica se conoce como análisis de series de Fourier . Para determinar la serie de Fourier, la señal de interés se correlaciona con sinusoides en frecuencias armónicamente relacionadas. Esta correlación proporciona la amplitud y la fase de una serie sinusoidal que representa la señal periódica, y se pueden graficar frente a la frecuencia para describir la composición del dominio de frecuencia de la señal.

La transformada de Fourier de la onda sinusoidal es $F [ s i n ω0t ] = - j π[ d( ω -ω0) - d( ω +ω0) ]$

# Espectro de frecuencia 
El espectro de frecuencia de una señal es el rango de frecuencias contenidas por una señal.
Los espectros ofrecen información que no aparece en el dominio del tiempo. La gráfica de frecuencias ofrece una representación rápida de la estructura de armónicas, que son como las huellas dactilares que ayudan a caracterizar y entender la forma de una señal complicada.

Para ejemplificar el espectro de frecuencia de una señal senoidal se presenta las siguientes graficas. La primera grafica presenta la señal senoidal mientras que la segunda el espectro de frecuencia de la señal senoidal dada de como se vería en un espectrometro.
Para poder apreciar el espectro de frecuecia de la señal senoidal en distinto tiempo, solo basta modificar el siguiente valor de la siguiente linea de codigo _signalTime = numpy.arange(1, **100**, 0.5)_, este valor de 100 puede ser cualquiera que el usuario desee, y a partir de este poder apreciar la segunda grafica del espectro de frecuencia.         
Fuente: [www.sciencedirect.com](https://www.sciencedirect.com/topics/engineering/frequency-spectrum)

<center><img src=IMG\EF.png></center>

```python
import numpy 
from matplotlib import pyplot  
   
signalTime = numpy.arange(1, 100, 0.5); 
  
signalAmplitude = numpy.sin(signalTime) 
  
pyplot.plot(signalTime, signalAmplitude, color ="green") 
  
pyplot.xlabel("Time") 
pyplot.ylabel("Amplitude") 
pyplot.title("Signal") 
pyplot.show() 
  
pyplot.magnitude_spectrum(signalAmplitude, color ="green") 
  
pyplot.title("Magnitude Spectrum of the Signal") 
pyplot.show() 
```
