# Explicación de las propiedades de las señales senoidales

### Introducción

Las señales senoidales o señales sinusoidales son patrones de ondas periodicas que matemáticamente pueden ser descritas mediante las funciones del seno y coseno. Estas describen acertadamente eventos naturales y señales variables en el tiempo.  
Este tipo de señales están compuestas por varias partes, las cuales se explicarán más adelante.

Se puede representar como:  
$y(t)=Asin(2πft+φ)$

Donde:
- $A$. Amplitud: indica el valor máximo a la que la senoide va a llegar. 
- $f$. Frecuencia: representa el ángulo girado por unidad de tiempo.
- $t$. Periodo: es la cantidad de tiempo en el que se acompleta un ciclo. *Esta es la variable independiente*.
- $φ$. Fase: es el desplazamiento a la izquiero o derecha.


[fuente](https://www.lifeder.com/onda-senoidal/)

```python
import matplotlib.pyplot as plt
import numpy as np

#* Variables que determinan los valores de la señal *#
amplitud = 1
frecuencia = 1
periodo = 1 / frecuencia
fase = 1

#* Frecuencia, periodo de muestreo y ciclos *#
frecuencia_Muestreo = 30 * frecuencia
periodo_Muestreo = 1 / frecuencia_Muestreo
ciclos = 5

t = np.arange( 0, ciclos * periodo, periodo_Muestreo ) # Variable independiente
x = amplitud * np.sin( 2 * np.pi * frecuencia * t + fase ) # Caulculo de la señal

#* Graficación de la señal *#
plt.axhline(y = 0, color = 'k')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.plot(t,x)
plt.grid('on')
```

# Explicación de una señal periódica

Las señales periódicas se caracterizan porque se les puede encontar un patrón que se repite constantemente cada determinado tiempo. El **periodo** es el tiempo que tarda una función en completar un ciclo y este se expresa en segundos. 

Para comprender porque una señal senoidal se considera periódica podríamos imaginar un plano y una circunferencia con su centro en el origen y radio de 1 unidad. Ahora supongamos que tenemos un punto móvil que se desplaza por el perimetro de la cirunferencia, y además, se traza un segmento desde el origen a este punto, de esta forma se genera un ángulo entre el radio y este segmento. Si calculamos el seno del ángulo mientras que desplazamos el punto móvil nos daremos cuenta que el resultado del seno se repetirá cada que se recorra el circulo completo.

Para poder visualizar el periodo en la gráfica, si el parametro **_"period"_** de la señal senoidal se modifica a 1, la forma de la señal se repetirá cada segundo, si se modifica a 2 se repetirá cada 2 segundos. Inténtalo.


<center><img src=https://www.universoformulas.com/imagenes/matematicas/trigonometria/grafica-seno.jpg></center>

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)


period=6.
x_max = 2*pi
x_min = 0
step = 0.01

# Configuración del gráfico visual
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [sine_wave(1,1/period, 0, time) for time in x]
plt.title('Señal senoidal con periodo modificable')
plt.plot(x,y)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
plt.grid('on')
a =[0,period]
b =[0,0]
plt.plot(a, b, linestyle="-", marker="o")
plt.text(period-1, .025, 'Periodo', color='orange')
plt.show()
```

<h1> Frecuencia </h1>

<p>Esta idea coincide con el concepto de periodo de una función periódica, como vimos, la función f(x)= sen x es periódica de periodo 2π y si consideras x0 ∈ R y el punto (x0, senx0)en su gráfica, como una partícula moviéndose a lo largo de la curva. <br> 
Ahora bien, observa que la partícula, en su movimiento a lo largo de un periodo, pasa por todos los posibles valores de la función, es decir completa un ciclo. Al número de ciclos completados en el intervalo [0,2π] se le llama velocidad angular, y se denota por la letra griega ω. O sea que la velocidad angular es el número de veces que se repite un patrón completo de una curva senoidal en el intervalo [0,2π], en la expresión general de la familia de funciones senoidales, corresponde al parámetro b. También se le conoce como pulsación (o simplemente pulso). <br>
De acuerdo con la definición de periodo, se tiene que T= 2π/ω <br>
Definimos la frecuencia de oscilación, fr, de una función senoidal como el recíproco del periodo, es decir fr = ω/2π = 1/T <br>
Sucede entonces que si la frecuencia es mayor, el periodo disminuye y viceversa.Su unidad es el hertzio, de símbolo Hz.</p>

<p> Para visualizar el concepto de frecuencia se realizo la grafica de dos señales senoidales. <br>
La primera señal esta definida por una frecuencia de valor 4, esta dada por la señal de color azul. <br>
La segunda grafica esta definida por una frecuencia de 10, esta dada por la señal de color naranja. <br>
Lo que se puede apreciar es que la segunda señal realiza un mayor numero de ciclos completados en un 1s o en el intervalo [0,2π]. <br>
Se puede modificar el valor de las frecuencias para poder apreciar su comportamiento, el valor de la frecuencia de la primera señal lo tiene la variable frecuencia1 y la de la segunda señal esta dada por la variable frecuencia2, estos valores se pueden modificar con cualquier valor, siempre y cuando la frecuencia2 sea mayor a frecuencia1</p>

```python
import numpy as np
import matplotlib.pyplot as plt
print("Frecuencia")
#señal1 
frecuencia1 = 4
Periodo1= 1/frecuencia1
A1= 2
theta1 = 0
#señal2
frecuencia2 = 10 #Tiene que ser mayor que la frecuencia1
Periodo2= 1/frecuencia2
A2= 2
theta2 = 0

Fs = 30*frecuencia2
Ts = 1/Fs
Nc= 4
plt.title("Señal senoidal con frecuencia modificable")
t = np.arange(0, Nc*Periodo1, Ts)
x = A1*np.sin(2*np.pi*frecuencia1*t+theta1)
x2 = A2*np.sin(2*np.pi*frecuencia2*t+theta2)
plt.plot(t,x,t,x2)
plt.grid('on')
```

# Señal compuesta
Una señal compuesta está compuesta por varias señales senoidales de diferentes aplitutes, frecuencias y fases. Pueden ser periódica o no periódica. Una señal compuesta periódica puede descomponerse en una serie de ondas senoidales simples con frecuencias discretas, frecuencias que tienen valores enteros. Mientras que una señal compuesta no periódica puede descomponerse en una combinación de un número infinito de ondas sinusoidales simples con frecuencias continuas, frecuencias que tienen valores reales.  
Un ejemplo de este tipo de señales es una señal de video, que tiene dos señales de audio y una señal de imagen.
